## Missing Values

```python
housing.isnull().sum()
```

Loop to flag missing values that can be applied across different DataFrames.

```python
# looping over columns with missing values
for col in housing:

    # creating columns with 1s if missing and 0 if not
    if housing[col].isnull().astype(int).sum() > 0:
        housing['m_'+col] = housing[col].isnull().astype(int)


# displaying the first five rows of the DataFrame
housing.head()
```

### Explore Missing Data Variables
Create a DataFrame after dropping missing values to more easily develop imputation strategies. Remember, some data analysis techniques do not work when missing values are present.

```python
# creating a dropped dataset to visualize 'Mas Vnr Area'
df_dropped = housing.dropna()


# displaying the plot for 'Mas Vnr Area'
sns.distplot(df_dropped["Mas Vnr Area"])
plt.show()
```

 - 0 inflated poisson distribution with left skewness
 - Hypothesis that exterior improvements increase the value of a home.
 - Since 0 is the main missing value we can use this as a gauge to fill NAs

```python
# Fill in missing values for 'Mas Vnr Area' with an appropriate value.

# creating an imputation value
fill = 0

# imputing 'Mas Vnr Area'
housing['Mas Vnr Area'] = housing['Mas Vnr Area'].fillna(fill)
```

```python
#Impute all other missing values in the dataset with the median for their respective columns
fill = housing['Total Bsmt SF'].median
housing['Total Bsmt SF'] = housing['Total Bsmt SF'].fillna(fill)


fill = housing['Garage Cars'].median
housing['Garage Cars'] = housing['Garage Cars'].fillna(fill)


fill = housing['Garage Area'].median
housing['Garage Area'] = housing['Garage Area'].fillna(fill)
```

```python
# making sure all missing values have been taken care of
housing.isnull().any().sum()
```
